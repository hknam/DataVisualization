```python
import csv
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = (15, 13))
plt.ylim(740, 1128)
plt.xlim(1965, 2011)
```

```python
with open('./mortality1.csv') as csvfile:
    mort_data = [row for row in csv.DictReader(csvfile)]
```

```python
x = []
males_y = []
females_y = []
every_y = []
```

```python
for row in mort_data:
    x += [int(row['Year'])]
    males_y += [int(float(row['Males']))]
    females_y += [int(float(row['Females']))]
    every_y += [int(float(row['Everyone']))]
```

```python
%matplotlib inline
plt.plot(x, males_y, color = '#1a51c3', label = 'Males', linewidth = 1.8)
plt.plot(x, females_y, color = '#bc108d', label = 'Female', linewidth = 1.8)
plt.plot(x, every_y, color = '#747e8a', label = 'Everyone', linewidth = 1.8)
plt.legend(loc = 0, prop = {'size' : 10})
plt.xlabel('Year', fontsize = 10)
plt.ylabel('Mortality', fontsize = 10)
plt.show()
```
