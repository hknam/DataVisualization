```python
import matplotlib.pyplot as plt
import csv
import operator
import datetime as dt
```

```python
with open('./ebola_data_db_format.csv', 'rt') as f:
    filtered_data = [row for row in csv.reader(f) if row[3] != '0.0'
                     and row[3] != '0' and 'deaths' in row[0]]
```

```python
sorted_data = sorted(filtered_data, key = operator.itemgetter(1))
```

```python
guinea_data = [row for row in sorted_data if row[1] == 'Guinea' and
              row[0] == 'Cumulative number of confirmed, probable and suspected Ebola deaths']
```

```python
sierra_data = [row for row in sorted_data if row[1] == 'Sierra Leone' and
              row[0] == 'Cumulative number of confirmed, probable and suspected Ebola deaths']
```

```python
liberia_data = [row for row in sorted_data if row[1] == 'Liberia' and 
               row [0] == 'Cumulative number of confirmed, probable and suspected Ebola deaths']
```

```python
g_x = [dt.datetime.strptime(row[2], '%Y-%m-%d').date() for row in guinea_data]
```

```python
g_y = [int(float(row[3])) for row in guinea_data]
```

```python
s_x = [dt.datetime.strptime(row[2], '%Y-%m-%d').date() for row in sierra_data]
```

```python
s_y = [int(float(row[3])) for row in sierra_data]
```

```python
l_x = [dt.datetime.strptime(row[2], '%Y-%m-%d').date() for row in liberia_data]
```

```python
l_y = [int(float(row[3])) for row in liberia_data]
```

```python
plt.figure(figsize = (10, 10))
plt.plot(g_x, g_y, color = 'red', linewidth = '2', label = 'Guinea')
plt.plot(s_x, s_y, color = 'orange', linewidth = '2', label = 'Sierra Leone')
plt.plot(l_x, l_y, color = 'blue', linewidth = '2', label = 'Liberia')
```

```python
plt.xlabel('Date', fontsize = 18)
plt.ylabel('Number of Ebola Deaths', fontsize = 18)
```

```python
plt.title('Probable and Suspected Ebola Deaths', fontsize = 20)
```

```python
plt.show()
```
