```python
import scipy as sp
from scipy.interpolate import splprep, splev
```

```python
def multiply_poly():
    cubic_1 = sp.poly1d([3, 4, 5, 5])
    cubic_2 = sp.poly1d([4, 1, -3 ,3])
    
    print(cubic_1)
    print(cubic_2)
    
    print('-' * 36)
    
    print(cubic_1 * cubic_2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0, 2.5, .1)
x = np.sin(2 * np.pi * t)
y = np.cos(2 * np.pi * t)
```

```python
tcktuples, uarray = splprep([x, y], s = 0)
```

```python
unew = np.arange(0, 1.01, 0.01)
```

```python
spline_values = splev(unew, tcktuples)
```

```python
plt.figure(figsize = (10, 10))
plt.plot(x, y, 'x', spline_values[0], spline_values[1], np.sin(2*np.pi * unew), np.cos(2*np.pi*unew), x, y, 'b')
```

```python
plt.legend(['Linear', 'Cubic Spline', 'True'])
plt.axis([-1.25, 1.25, -1.25, 1.25])
plt.title('Parametric Spline Interpolation Curve')
```

```python
plt.show()
```
