```python
import numpy as np
```

```python
from time import time
```

```python
def incrembyone(x):
    return x + 1
```

```python
data_array = np.linspace(1, 5, 1000000)
```

```python
t1 = time()
```

```python
len_data = len(data_array)
```

```python
print('Len = ', str(len_data))
```

```python
print(data_array[1:7])
```

```python
for i in range(len_data):
    data_array[i] += 1
```

```python
print('time for loop (No vectorization) -> '+ str(time() - t1))
```

```python
t2 = time()
```

```python
vecincr = np.vectorize(incrembyone)
vecincr(data_array)
```

```python
print('time for vectorized version-1 : ' + str(time() - t2))
```

```python
t3 = time()
```

```python
data_array += 1
```

```python
print(data_array[1:7])
```

```python
print('time for vectorized version=2 : ' + str(time() - t3))
```
