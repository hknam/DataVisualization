```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
students = pd.read_csv('./ucdavis.csv')
```

```python
grid = sns.FacetGrid(students, hue='gender', palette = 'Set1', size = 6)
```

```python
grid.map(plt.scatter, 'gpa', 'computer', s = 250, linewidth = 0.65, edgecolor = 'white')
```

```python
grid.add_legend()
```

```python
plt.show()
```
